```python
train1['combined']=train1.apply(lambda x:'%s %s %s %s' % (x['make'],x['model'],x['variant'],x['fuel_type']),axis=1)
master['combined']=master.apply(lambda x:'%s %s %s %s' % (x['make'],x['model'],x['variant'],x['fuel_type']),axis=1)
```

```python
train1_combined = []
master_combined = []

for i in train1['combined']:
    train1_combined.append(i)
    
for i in master['combined']:
    master_combined.append(i)
    
```

```python
max_ratios = []
for i in range(len(train1_combined)):
    max = 0
    match_pos = 0
    for j in range(len(master_combined)):
#         match_pos = 0
        
        t1 = train1_combined[i]
        t2 = master_combined[j]
#         print(fuzz.ratio(t1,t2))
        if fuzz.ratio(t1,t2) > max:
            max = fuzz.ratio(t1,t2)
            match_pos = j
#             print(j)
    max_ratios.append(match_pos)
        
```
